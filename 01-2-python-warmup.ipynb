### The Collatz Conjecture (aka the 3n+1 problem)

The Collatz conjecture is a famous unsolved problem in mathematics.  The conjecture states that given an arbitrary positive integer the following operation will terminate with 1 after a finite number of steps:

```
     if n is even:
         n = n // 2
     else:
         n = 3*n+1
```
Wikipedia has a short article on the Collatz conjecture.  Though simple to state and comprehend this turns out to be a very hard conjecture to prove.  Paul Erd≈ës said about the Collatz conjecture: "Mathematics may not be ready for such problems." Jeffrey Lagarias in 2010 claimed that based only on known information about this problem, "this is an extraordinarily difficult problem, completely out of reach of present day mathematics."

The sequence of numbers produced by the above operation are known as hailstone numbers as the values go up and down until they reach a certain size (power of 2) and then rapidly "fall down" to 1 like real hailstones.

The number of times we iterate before n becomes 1 is known as the _stopping time_

````python
# we can define an empty function in Python with the 'pass' construct

def stopping_time(n):
    pass 
````

````python
# In Python, functions are first class.  
# So if we evaluate the name of a function, its value is the function

stopping_time 
````

````python
# A simple definition

def stopping_time(n):
    cnt = 0
    while n > 1:
        cnt += 1
        if n%2 == 0:
            n = n // 2
        else:
            n = 3*n+1  
    return cnt
````

````python
stopping_time(11)
````

## List Comprehensions

Consider the following template where we initialize a variable with an empty list and over each iteration of a loop we append a value:

````python
# Calculate the squares of a list of numbers
lst = []
for i in [1,2,3,4]:
    lst.append(i*i)
lst
````

The same computation can be expressed concisely using a list comprehension

````python
[  i*i  for i in [1,2,3,4] ]
````

The general form of a list comprehension has 3 parts:

```
[  <expression>   <for loop>    <test>  ]
```

A list comprehension to find the squares of the even numbers less than 10.  

````python
[ i*i  for i in range(10) if i%2==0]
````

Computer the stopping times for all integers in the range [1,10]

````python
[ ( i, stopping_time(i))  for i in range(1,11)]
````

Lets plot the stopping time of various integers

````python
import matplotlib.pyplot as plt 
%matplotlib inline 
````

````python
x = range(1, 100)
y = [stopping_time(n) for n in x]
p = plt.plot(x,y)
````

The sequence of numbers we get when performing the `3n+1` operation is known as the Collatz sequence (aka hailstone numbers).

````python
def collatz_sequence(n):
    lst = []
    while n > 1:
        lst.append(n)
        if n%2 == 0:
            n = n // 2
        else:
            n = 3*n+1
    lst.append(1)
    return lst
````

````python
collatz_sequence(24)
````

````python
n = 2500
y = collatz_sequence(n)
x = range(0, len(y)) 
plt.plot(x,y)
plt.plot()
````

## More graphs

A scatter plot of various stopping times (similar to the graph in the Wikipedia article)

````python
x = range(1,10000)
y = [stopping_time(n) for n in x]
plt.scatter(x,y, c='r', marker='.')
plt.plot()
````
